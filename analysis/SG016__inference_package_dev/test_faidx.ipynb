```python
fa = pyfastx.Fasta('GRCh38_no_alt_analysis_set_GCA_000001405.15.fasta.gz')
```

```python
for i,seq in enumerate(fa):
    print(f'field {i}')
    print(f'\tname field:\n\t\t{seq.name}')
    print(f'\tseq field length:\n\t\t{len(seq.seq)}')
    print(f'\tdescription field:\n\t\t{seq.description}')
    print('')
```
