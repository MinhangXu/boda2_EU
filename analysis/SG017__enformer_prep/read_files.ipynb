```python
outgoing_data= zip(
    mpra_table['IDs_ref'], 
    mpra_table['K562_mean_ref'],
    mpra_table['K562_mean_alt'],
    mpra_table['K562_pred_aggreg_ref'],
    mpra_table['K562_pred_aggreg_alt'],
    mpra_table['HepG2_mean_ref'],
    mpra_table['HepG2_mean_alt'],
    mpra_table['HepG2_pred_aggreg_ref'],
    mpra_table['HepG2_pred_aggreg_alt'],
    mpra_table['SKNSH_mean_ref'],
    mpra_table['SKNSH_mean_alt'],
    mpra_table['SKNSH_pred_aggreg_ref'],
    mpra_table['SKNSH_pred_aggreg_alt']
)

filtered_data = [ x for x in outgoing_data if len(x[0].split(':')) == 6 ]
```

```python
filtered_data[0]
len(filtered_data)
```

```python
with open('quick_vars.vcf', 'w') as f:
    print('\t'.join(['chrom', 'pos', 'id', 'ref', 'alt', 'info']), file=f)
    for i, info in enumerate(filtered_data):
        ids, *y = info
        data = ';'.join([str(x) for x in y])
        chrom, pos, ref, alt, value, tag = ids.split(':')
        print( '\t'.join([chrom, pos, ids, ref, alt, data]), file=f )
```
