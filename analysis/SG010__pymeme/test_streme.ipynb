```python
import os
import sys

boda_src = os.path.join( os.path.dirname( os.path.dirname( os.getcwd() ) ), 'src' )
sys.path.insert(0, boda_src)

from pymeme import streme, parse_streme_output
```

# Streme on a file

```python
test_seq_file = '/home/ubuntu/boda2/analysis/SG010__pymeme/streme_test.txt'
```

```python
streme_results = streme(test_seq_file)
for i, line in enumerate(streme_results['output'].decode("utf-8").split('\n')):
    print(line)
```

# Streme on a list of sequences

```python
seqs_as_list = []
with open(test_seq_file,'r') as f:
    for line in f:
        if '>' != line[0]:
            seqs_as_list.append(line.rstrip())
            
seqs_as_list
```

```python
streme_results = streme(seqs_as_list)
for i, line in enumerate(streme_results['output'].decode("utf-8").split('\n')):
    print(line)
```

# Streme on a dictionary of sequences

```python
seqs_as_dict = {}
with open(test_seq_file,'r') as f:
    for line in f:
        if '>' == line[0]:
            key = line.rstrip()
        else:
            seqs_as_dict[key] = line.rstrip()
            
seqs_as_dict
```

```python
streme_results = streme(seqs_as_dict)
for i, line in enumerate(streme_results['output'].decode("utf-8").split('\n')):
    print(line)
```

# Parse Streme output

```python
parse_streme_output(streme_results['output'])
```

```python
parse_streme_output(streme_results['output'])['motif_results'][0]['ppm'].shape
```
